```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Node = pd.read_csv('./data/SDGE_NODE_LMP/SANYSIDRO_N001_0813_0815.csv')
Node
```

```python
# Get LMP Online
LMP = Node[Node['LMP_TYPE']=='LMP'] 
MCC = Node[Node['LMP_TYPE']=='MCC'] # Congestion
MCE = Node[Node['LMP_TYPE']=='MCE'] # Energy
MCL = Node[Node['LMP_TYPE']=='MCL'] # Loss

LMP
```

```python
price = LMP['PRC'].to_numpy()
congestion = MCC['PRC'].to_numpy()
energy = MCE['PRC'].to_numpy()
loss = MCL['PRC'].to_numpy()

time = LMP['INTERVALSTARTTIME_GMT'].to_numpy()
```

```python
plt.plot(time,price)
plt.xlabel('Time') 
plt.ylabel('Price') 
  
# displaying the title
plt.title("LMP")
```

```python
plt.plot(time,congestion)
plt.xlabel('Time') 
plt.ylabel('Price') 
plt.xticks(rotation=90)
  
# displaying the title
plt.title("MCC")
```

```python
plt.plot(time,energy)
plt.xlabel('Time') 
plt.ylabel('Price') 
  
# displaying the title
plt.title("MCE")
```

```python
plt.plot(time,loss)
plt.xlabel('Time') 
plt.ylabel('Price') 
  
# displaying the title
plt.title("MCL")
```
