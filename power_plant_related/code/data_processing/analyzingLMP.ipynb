```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
date = '0901_0905'
mcc = pd.read_csv('./SDGE_Node_MCC_TS_T_{}.csv'.format(date))
lmp = pd.read_csv('./SDGE_Node_LMP_TS_T_{}.csv'.format(date))

lmp_bins = np.arange(50,1000,1).tolist()
mcc_bins = np.arange(-1000,2000,0.5).tolist()

def preprocessing(the_lmp,bins,idx):
    N = the_lmp.shape[0]
    print('lmp entries: {}'.format(N))
    time = the_lmp.iloc[idx].to_numpy()[0]
    np_lmp = the_lmp.iloc[idx].to_numpy()[1:].tolist()
    np_lmp = np.asarray(np_lmp,dtype=float)

    
    plt.hist(np_lmp, bins = bins) 
    plt.title("{}".format(time)) 
    plt.show()
    return np_lmp,N

np_lmp,n_lmp = preprocessing(lmp,lmp_bins,500)
np_mcc,n_mcc = preprocessing(mcc,mcc_bins,500)
```

```python
mcc = mcc.sort_values(by=['Unnamed: 0']).iloc[:-1]
mcc
```

```python
lmp = lmp.sort_values(by=['Unnamed: 0']).iloc[:-1]
lmp
```

```python
import matplotlib.animation as animation
```

```python
def prepare_animation(bar_container):
    def animate(frame_number):
        # simulate new data coming in
        time = lmp.iloc[frame_number].to_numpy()[0]
        np_lmp = lmp.iloc[frame_number].to_numpy()[1:].tolist()
        np_lmp = np.asarray(np_lmp,dtype=float)

        ax.set_title('Date: {}'.format(time))
        ax.set_xlabel("Price")
        ax.set_ylabel("Count")

        n, _ = np.histogram(np_lmp, lmp_bins)
        for count, rect in zip(n, bar_container.patches):
            rect.set_height(count)
        return bar_container.patches
    return animate

# def generateVideo(np_lmp,name,bins,N):    
    # histogram our data with numpy
name = 'LMP'
n, _ = np.histogram(np_lmp, lmp_bins)

fig, ax = plt.subplots()
_, _, bar_container = ax.hist(np_lmp, lmp_bins, lw=1,
                              ec="yellow", fc="green", alpha=0.5)
ax.set_ylim(top=170)  # set safe limit to ensure that all data is visible.

ani = animation.FuncAnimation(fig, prepare_animation(bar_container), n_lmp-1,
                              repeat=True, blit=False)

writervideo = animation.FFMpegWriter(fps=5)
ani.save('./{}_TS_Histogram_{}.mp4'.format(name,date), writer=writervideo)
plt.close()
#     print('DONE')
```
