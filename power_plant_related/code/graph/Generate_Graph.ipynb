```python
def add_connection(df,Node1,Node2,volt):
    df.loc[(Node1,Node2)] = volt
    df.loc[(Node2,Node1)] = volt
    return df

def del_connection(df,Node1,Node2):
    df.loc[(Node1,Node2)] = 0
    df.loc[(Node2,Node1)] = 0
    return df
```
