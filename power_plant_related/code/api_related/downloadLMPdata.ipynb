```python
import requests
import numpy as np
import pandas as pd
import time
```

```python
links = pd.read_csv('CAISE_SD_NODES_API_LINK_0901-0905.csv')
nodes = links['PNODE_ID'].to_numpy()
links_list = links['LINK'].to_numpy()
N = nodes.shape[0]

for i in range(N):
    print('Downloading {}/{} {} started ---> '.format(i,N,nodes[i]),end=' ')
    req = requests.get(links_list[i])
    time.sleep(5)
    
    with open('../.././data/SDGE_NODE_LMP/{}.zip'.format(nodes[i]),'wb') as output_file:
        output_file.write(req.content)
    
    print('Downloading Completed')
    
```
